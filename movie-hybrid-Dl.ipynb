```python
import pandas as pd
import numpy as np

# Step 1: Calculate the average rating across all movies
v_avg = metadata_df['vote_average'].mean()

# Step 2: Define the weighted rating function
def weighted_rating(x, m, v_avg):
    v = x['vote_count']
    R = x['vote_average']
    
    # Apply the weighted rating formula
    return (v / (v + m) * R) + (m / (m + v) * v_avg)

# Step 3: Set a minimum vote count threshold (m)
m = 1000  # You can adjust this threshold based on your data

# Step 4: Filter movies with a sufficient number of votes (optional)
metadata_df = metadata_df[metadata_df['vote_count'] >= m]

# Step 5: Apply the weighted rating function to the metadata dataframe and create the new column
metadata_df['weighted_rating'] = metadata_df.apply(weighted_rating, axis=1, m=m, v_avg=v_avg)
```

# Content Base Filtering

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer

# Function to handle cases where the value might be a list
def to_string(value):
    if isinstance(value, list):
        return ' '.join(value)
    elif isinstance(value, str):
        return value
    else:
        return str(value)  # Convert other types to string if necessary

# Preprocessing: Combine relevant textual columns into a single feature
smd['combined_features'] = smd['overview'].fillna('').apply(to_string) + " " + \
                            smd['keywords'].fillna('').apply(to_string) + " " + \
                            smd['genres'].apply(to_string)

# Initialize a TF-IDF Vectorizer
vectorizer = TfidfVectorizer(stop_words='english')

# Fit and transform the combined features
tfidf_matrix = vectorizer.fit_transform(smd['combined_features'])

# Now you can proceed with creating a content-based filtering system using this tfidf_matrix
```

```python
smd
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get recommendations for a given movie name
def get_recommendations_by_name(movie_name, cosine_sim=cosine_sim, top_n=5):
    # Find the index of the movie based on its name
    movie_idx = smd[smd['title'] == movie_name].index[0]
    
    # Get the pairwise similarity scores for the given movie
    sim_scores = list(enumerate(cosine_sim[movie_idx]))
    
    # Sort the movies based on similarity scores in descending order
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Get the top N most similar movies (excluding the movie itself)
    sim_scores = sim_scores[1:top_n+1]
    
    # Get the indices of the most similar movies
    movie_indices = [i[0] for i in sim_scores]
    
    # Retrieve the movie titles for the recommended movies
    recommended_titles = smd['title'].iloc[movie_indices]
    
    # Format the recommendation message
    recommendation_message = f"Recommended movies for the movie '{movie_name}' are:\n"
    recommendation_message += '\n'.join(recommended_titles)
    
    return recommendation_message

# Example: Get recommendations for the movie "The Matrix"
movie_name = "Jumanji"
recommendation_message = get_recommendations_by_name(movie_name)
print(recommendation_message)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Load the smd dataset Replace with the path to your smd dataset

# Fill NaN values with empty strings in text columns
smd['combined_features'] = smd['combined_features'].fillna('')

# Calculate TF-IDF matrix based on combined features (cast, genres, keywords, etc.)
tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(smd['combined_features'])

# Compute the cosine similarity matrix based on combined features
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Create a Series to map movie titles to indices
indices = pd.Series(smd.index, index=smd['original_title']).drop_duplicates()

# Recommendation function based on content similarity and weighted ranking
def get_recommendations(title, cosine_sim=cosine_sim, n=10):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort movies based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Select the top N similar movies
    sim_scores = sim_scores[1:n+1]  # Exclude the movie itself

    # Get movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Filter and rank movies by weighted_rating and vote_count for quality control
    recommendations = smd.iloc[movie_indices].sort_values(by=['weighted_rating', 'vote_count'], ascending=False)
    return recommendations[['original_title', 'weighted_rating', 'vote_count']]

# Example: Get recommendations for a specific movie
recommended_movies = get_recommendations("Jumanji", n=10)
print("Recommended movies:\n", recommended_movies)
```

# Hybrid DL with content and collaborative Approach

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split as surprise_train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.metrics import mean_squared_error

# Load your smd and ratings_df dataframes
# smd = pd.read_csv("path/to/smd.csv")
# ratings_df = pd.read_csv("path/to/ratings.csv")

# Fill NaN values in the combined features column
smd['combined_features'] = smd['combined_features'].fillna('')

# Calculate TF-IDF matrix for content-based filtering
tfidf = TfidfVectorizer(stop_words='english')
content_features = tfidf.fit_transform(smd['combined_features']).toarray()

# Collaborative Filtering using SVD
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings_df[['userId', 'movieId', 'rating']], reader)
trainset, testset = surprise_train_test_split(data, test_size=0.2)

# Train the SVD model
svd = SVD(n_factors=20)
svd.fit(trainset)

# Create SVD features for movies
smd['svd_features'] = smd['id'].apply(
    lambda x: svd.qi[trainset.to_inner_iid(x)] if x in trainset._raw2inner_id_items else np.zeros(20)
)

# Combine content and collaborative features
combined_features = np.hstack((content_features, np.vstack(smd['svd_features'].values)))

# Define labels as average ratings for each movie
movie_avg_ratings = ratings_df.groupby('movieId')['rating'].mean()
smd['avg_rating'] = smd['id'].map(movie_avg_ratings)
labels = smd['avg_rating'].fillna(smd['avg_rating'].mean()).values  # Handle NaN values

# Standardize the features
scaler = StandardScaler()
combined_features_scaled = scaler.fit_transform(combined_features)

# Split combined features and labels into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    combined_features_scaled, labels, test_size=0.2, random_state=42
)

# Define the Sequential model with Dense and Dropout layers
model = Sequential()

# Input layer + First hidden layer
model.add(Dense(128, input_dim=X_train.shape[1], activation='relu'))
model.add(Dropout(0.2))  # Dropout with 20% rate

# Second hidden layer
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(1))  # For regression (single output)

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Evaluate the model on the test set
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print(f"Mean Squared Error (MSE): {mse}")


```

```python
# Recommendation function based on combined features
def get_combined_recommendations(title, n=10):
    # Get the index of the movie that matches the title
    idx = smd[smd['original_title'] == title].index[0]
    
    # Calculate similarity with other movies
    movie_features = combined_features[idx].reshape(1, -1)
    similarities = cosine_similarity(movie_features, combined_features).flatten()
    
    # Get indices of top recommendations
    similar_indices = np.argsort(similarities)[-n-1:-1][::-1]  # Exclude the movie itself
    
    # Retrieve the recommended movies
    recommended_movies = smd.iloc[similar_indices]
    return recommended_movies[['original_title', 'weighted_rating', 'vote_count']]
```
